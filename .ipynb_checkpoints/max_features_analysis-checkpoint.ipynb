# Max features parameter

```python
import os
from scipy.io import arff
import pandas as pd
import numpy as np
import math
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load data
rf_sklearn = pd.DataFrame(arff.loadarff('data/meta_random_forest_100_1000max.arff')[0])
rf_sklearn['max_features_int'] = np.ceil(rf_sklearn['max_features'] * rf_sklearn['NumberOfFeatures'])

rf_ranger = pd.read_csv('data/meta_ranger.csv')
rf_ranger['max_features_int'] = rf_ranger['mtry'] 

rf_all = {'sklearn': rf_sklearn,
          'ranger': rf_ranger}
```

```python
# Get top n hyperparameter settings
n = 10 # retrieve only n settings per dataset per classifier with highest accuracy  
m = 40 # minimum number of observations per dataset to be included

rf_topn = {}
for name, rf_df in rf_all.items():
    groups = []
    for g in rf_df.groupby(['task_id']):
        group = pd.DataFrame(g[1])
        group['task_id'] = g[0]
        if len(group) > m:
            groups.append(group.sort_values(by = 'y', ascending  = False)[0:n])
        else: 
            print('Dataset %s not included for classifier %s because no. obervations is %s.' %(int(g[0]), alg, int(len(group))))
    rf_topn[name] = pd.concat(groups)
    print('Nr. of observations for %s: %s' % (name, len(rf_topn[name])))
```

# 1. Histograms

The ratios were sampled from a uniform distribution between 

* sklearn: 0.1 and 0.9
* ranger: 0.0 and 1.0

The absolute number of features was determined by taking ceil(ratio).

## 1.1 Ratio

For the sklearn data there seems to be a clear pattern. For the ranger data, this pattern is not visible, especially when the number of features becomes lower. This sparks the belief that the pattern in the sklearn data might be caused by the smaller ratio range.

```python
rows = 1
cols = 2

for filter_val in [2000, 500, 200]:
    fig = plt.figure(figsize = (12,4))
    index = 1
    for name in ['sklearn', 'ranger']:
        df_all = rf_all[name][rf_all[name]['NumberOfFeatures'] < filter_val]
        df_topn = rf_topn[name][rf_topn[name]['NumberOfFeatures'] < filter_val]
        plt.subplot(rows, cols, index)
        plt.hist(df_all['max_features_int']/df_all['NumberOfFeatures'], 
                 density = True, alpha = 0.7, label = 'all')
        plt.hist(df_topn['max_features_int']/df_topn['NumberOfFeatures'], 
                 density = True, alpha = 0.7, label = 'top n')
        index += 1
        plt.title(name)
        plt.xlabel('max_features ratio')
        plt.legend()
    fig.suptitle('x = max. no. features/no. features, filtered by no. features < %s' %filter_val, size = 14)
    plt.show()
```

# 2. Log-log plots

Plots of the (absolute) max. number of features versus the number of features.

* Green line: $\sqrt(n\_features)$
* Red line: $\log_2(n\_features)$

```python
for filter_val in [2000, 500, 200]:
    rf_topn_f = rf_topn['ranger'][rf_topn['ranger']['NumberOfFeatures'] < filter_val]
    g = sns.jointplot(np.log10(rf_topn_f['NumberOfFeatures']), np.log10(rf_topn_f['max_features_int']), 
                      kind = 'hex', gridsize = 20, cmap = 'viridis')
    g.ax_joint.set_xticklabels(['%.f'%x for x in 10**g.ax_joint.get_xticks()])
    g.ax_joint.set_yticklabels(['%.f'%x for x in 10**g.ax_joint.get_yticks()])
    
    x = range(int(min(rf_topn_f['NumberOfFeatures'])), int(max(rf_topn_f['NumberOfFeatures'])), 10)
    g.ax_joint.plot(np.log10(x), np.log10(np.sqrt(x)), '-', label = 'sqrt')
    g.ax_joint.plot(np.log10(x), np.log10(np.log2(x)), '-', label = 'log')
    plt.suptitle('Ranger, no. features < %s, no. datasets = %s' %(filter_val,len(rf_topn_f['task_id'].unique())))
    plt.show()
```

# 3. Functions
Fit functions and see which one fits the data best:

1. $max\_features = a * n\_features$ 
2. $max\_features = n\_features^a$
3. $max\_features = a ^{\sqrt{n\_features}}$

```python
from scipy.optimize import curve_fit

def func1(x, a):
    return a * x
def func2(x, a):
    return x ** a
def func3(x, a):
    return a**np.sqrt(x)

def evaluate(x, y, popt):
    residuals = y - func(x, popt)
    ss_res = np.sum(residuals**2)
    rmse = np.sqrt(ss_res/len(x))
    ss_tot = np.sum((y-np.mean(y))**2)
    r_squared = 1 - (ss_res/ss_tot)
    return rmse, r_squared

for name in ['ranger', 'sklearn']:
    for filter_val in [2000, 500, 200]:
        rf_topn_f = rf_topn[name][rf_topn[name]['NumberOfFeatures'] < filter_val]
        print("\n%s: < %s features (%s datasets)\n" % (name, filter_val, len(rf_topn_f['task_id'].unique())))
        x = rf_topn_f['NumberOfFeatures']
        y = rf_topn_f['max_features_int']
        for func in [func1, func2, func3]:
            popt, pcov = curve_fit(func, x, y)
            rmse, r_squared = evaluate(x, y, popt)
            if func == func1:
                print('\t m = %.2f*n,\t \t RMSE: %.2f, \t R2: %.2f'% (popt[0], rmse, r_squared))
            elif func == func2:
                print('\t m = n^%.2f,\t \t RMSE: %.2f, \t R2: %.2f'% (popt[0], rmse, r_squared))
            elif func == func3:
                print('\t m = %.2f^sqrt(n),\t RMSE: %.2f, \t R2: %.2f'% (popt[0], rmse, r_squared))
```
